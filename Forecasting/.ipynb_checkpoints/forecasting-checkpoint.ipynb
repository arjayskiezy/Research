```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Step 1: Load the dataset with necessary columns
data = pd.read_csv('./Dataset/forecasting.csv', usecols=['REF_DATE', 'VALUE'], dtype={'VALUE': 'float64'})

# Convert 'REF_DATE' to datetime
data['REF_DATE'] = pd.to_datetime(data['REF_DATE'], format='%Y-%m')

# Step 2: Filter for passenger car sales and unadjusted data (if other filtering is needed)
filtered_data = data.rename(columns={'REF_DATE': 'Date', 'VALUE': 'Sales'})

# Handle duplicate dates by aggregating sales
filtered_data = filtered_data.groupby('Date')['Sales'].sum().reset_index()

# Set 'Date' as the index and ensure consistent monthly frequency
filtered_data.set_index('Date', inplace=True)
filtered_data = filtered_data.asfreq('MS')  # Set frequency explicitly

# Handle missing values using forward fill
filtered_data['Sales'] = filtered_data['Sales'].ffill()

# Step 3: Plot the time series
plt.figure(figsize=(12, 6))
plt.plot(filtered_data['Sales'], label='Monthly Passenger Car Sales')
plt.title('Monthly Passenger Car Sales (Canada)')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Step 4: Simple Moving Average (SMA)
filtered_data['SMA_12'] = filtered_data['Sales'].rolling(window=12).mean()

# Plot SMA
plt.figure(figsize=(12, 6))
plt.plot(filtered_data['Sales'], label='Actual Sales')
plt.plot(filtered_data['SMA_12'], label='12-Month SMA', linestyle='--')
plt.title('Simple Moving Average (12-Month)')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Step 5: Stationarity Check with ADF Test
def adf_test(series):
    result = adfuller(series)
    print("ADF Statistic:", result[0])
    print("p-value:", result[1])
    print("Critical Values:", result[4])
    if result[1] <= 0.05:
        print("The data is stationary.")
    else:
        print("The data is non-stationary.")

# Perform ADF test on original data
print("ADF Test on Original Sales Data:")
adf_test(filtered_data['Sales'])

# Step 6: Fit ARIMA Model (Skip differencing if data is stationary)
model = ARIMA(filtered_data['Sales'], order=(1, 1, 1))  # Adjust p, d, q based on ACF/PACF
arima_result = model.fit()

# Print ARIMA model summary
print("ARIMA Model Summary:")
print(arima_result.summary())

# Step 7: Forecasting with ARIMA
filtered_data['ARIMA_Forecast'] = arima_result.predict(start=len(filtered_data) - 12, 
                                                       end=len(filtered_data) + 12, 
                                                       dynamic=True)

# Plot ARIMA forecast vs actual
plt.figure(figsize=(12, 6))
plt.plot(filtered_data['Sales'], label='Actual Sales')
plt.plot(filtered_data['ARIMA_Forecast'], label='ARIMA Forecast', linestyle='--')
plt.title('ARIMA Model Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Step 8: Evaluate SMA and ARIMA Models
# SMA evaluation
sma_test = filtered_data['SMA_12'].dropna()[-12:]  # Last 12 months for SMA
sma_actual = filtered_data['Sales'][-len(sma_test):]

sma_mae = mean_absolute_error(sma_actual, sma_test)
sma_mse = mean_squared_error(sma_actual, sma_test)
sma_rmse = np.sqrt(sma_mse)

print("SMA Model Evaluation:")
print(f"MAE: {sma_mae}, MSE: {sma_mse}, RMSE: {sma_rmse}")

# ARIMA evaluation
arima_actual = filtered_data['Sales'][-12:]  # Last 12 months as test set
arima_pred = filtered_data['ARIMA_Forecast'][-12:]

arima_mae = mean_absolute_error(arima_actual, arima_pred)
arima_mse = mean_squared_error(arima_actual, arima_pred)
arima_rmse = np.sqrt(arima_mse)

print("ARIMA Model Evaluation:")
print(f"MAE: {arima_mae}, MSE: {arima_mse}, RMSE: {arima_rmse}")

# Step 9: Forecast for Next 12 Months
future_forecast = arima_result.get_forecast(steps=12)
forecast_df = future_forecast.summary_frame()

# Set forecast index to align with time series
forecast_df.index = pd.date_range(start=filtered_data.index[-1] + pd.DateOffset(months=1), 
                                  periods=12, freq='MS')

# Plot future forecast
plt.figure(figsize=(12, 6))
plt.plot(filtered_data['Sales'], label='Historical Data')
plt.plot(forecast_df['mean'], label='Forecast', linestyle='--')
plt.fill_between(forecast_df.index, forecast_df['mean_ci_lower'], forecast_df['mean_ci_upper'], 
                 color='gray', alpha=0.3)
plt.title('12-Month Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Step 10: Save Forecast Data
forecasted_data = pd.concat([filtered_data['Sales'], forecast_df['mean']], axis=1)
forecasted_data.columns = ['Historical Sales', 'Forecasted Sales']
forecasted_data.to_csv('forecasted_sales.csv')
print("Forecast saved to 'forecasted_sales.csv'.")
```
